## 🎯 Entrenamiento

### ✔ Modelo 1: Regresión Logistica

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)

y_pred_log_normal = log_model.predict(X_test)

print('Regresion Logistica')
print(classification_report(y_test, y_pred_log_normal))
```

###🔍 Resultados del Modelo Regresión Logistica

📌Clase 0 (No Deserción):

✅Precision ➡ 84%: Cuando el modelo predice que un cliente no desertará, es correcto el 84% de las veces.

✅Recall ➡ 90%: El modelo fue capaz de identificar correctamente el 90% de todos los clientes que realmente no desertaron.

✅F1-score ➡ 0.87: Un buen equilibrio entre precisión y recall para esta clase.

📌Clase 1 (Deserción):

Precision ➡ 65%: Cuando el modelo predice que un cliente sí desertará, solo es correcto el 65% de las veces.

🔴Hay un 35% de falsos positivos (clientes predichos como desertores que en realidad no lo fueron).

Recall ➡ (52%): El modelo solo pudo identificar correctamente el 52% de todos los clientes que realmente desertaron.

🔴Esto significa que el 48% de los clientes que sí desertaron fueron clasificados erróneamente como no desertores (falsos negativos).

🔴F1-score ➡ 0.58: Es considerablemente más bajo que el de la clase 0, reflejando el rendimiento más débil del modelo para la clase que deserta.

El modelo de Regresión Logística tiende a favorecer a la clase mayoritaria (no deserción).

Conclusión: El modelo actual es bastante bueno para predecir a los clientes que permanecerán, pero no es tan efectivo para identificar a los que van a desertar.

🔴Recall ➡ 47%: El modelo solo pudo identificar correctamente el 47% de todos los clientes que realmente desertaron. Esto es inferior al 52% de la Regresión Logística, lo que significa que Random Forest detecta menos desertores reales (más falsos negativos).

###✔ Modelo 4: Regresión Logística con balanceo y SMOTE

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE

X = datos.drop('Desercion', axis=1)
y = datos['Desercion']

X= pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

smote = SMOTE(random_state=42)
X_train_rest, y_train_rest = smote.fit_resample(X_train, y_train)

scaler = StandardScaler()
X_train_rest_scaled = scaler.fit_transform(X_train_rest)
X_test_scaled = scaler.transform(X_test)

log_model = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
log_model.fit(X_train_rest_scaled, y_train_rest)

y_pred = log_model.predict(X_test_scaled)

print('Regresión Logística con balanceo y SMOTE')
print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)
print('Matriz de confusión:\n', cm)
```

###🔍Resultados del Modelo:
1. 📌Clase 0 (No Deserción):

 ✅Precisión ➡ 88%:  Predice correctamente la mayoría de los clientes que no abandonan, aunque ligeramente menor que el 91% del modelo de Regresión Logística balanceado puro, pero mejor que el 84% del modelo logístico normal.

 🔴Recall ➡ 76%: El modelo identifica correctamente el 76% de todos los clientes que realmente no desertaron. Esto es un buen equilibrio, aunque inferior al 90% del modelo logístico normal y al 73% del balanceado puro.

2. 📌Clase 1 (Deserción):

 🔴Precisión ➡ 51%: Cuando el modelo predice que un cliente sí desertará, es correcto el 51% de las veces.
 Esta precisión es la más baja de los modelos probados, lo que significa que este modelo tiene la mayor cantidad de falsos positivos en la predicción de deserción. Es decir, predice que muchos clientes desertarán, pero aproximadamente la mitad de esas predicciones serán incorrectas.

 ✅Recall ➡ 71%: El modelo pudo identificar correctamente el 71% de todos los clientes que realmente desertaron. Esta es una mejora significativa en la capacidad de detectar desertores reales en comparación con los modelos no balanceados (52% en logístico normal y 47% en Random Forest), aunque ligeramente inferior al 79% del modelo balanceado puro.

📌Conclusión: este modelo es útil si tel objetivo es identificar la mayor cantidad posible de futuros desertores, incluso si eso significa tener más falsas alarmas (falsos positivos).
